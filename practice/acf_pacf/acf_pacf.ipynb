```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Sample data: let's simulate a time series
np.random.seed(42)
data = np.random.randn(100).cumsum()  # Random walk

# Plot ACF
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plot_acf(data, lags=20, ax=plt.gca())  # lags=20 means check correlation up to 20 steps
plt.title("ACF Plot")

# Plot PACF
plt.subplot(1,2,2)
plot_pacf(data, lags=20, ax=plt.gca(), method='ywm')  # method can be 'ywm', 'ols', etc.
plt.title("PACF Plot")

plt.tight_layout()
plt.show()
```
