```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.stats.diagnostic import acorr_ljungbox
```

```python
#generating random walk and white noise for comparison
np.random.seed(0)
n=1000

#white noise
white_noise=np.random.normal(0,1,n)

#random walk
random_shocks=np.random.normal(0,1,n)
random_walk=np.cumsum(random_shocks)

#plotting the series
plt.figure(figsize=(14,6))
plt.subplot(2,1,1)
plt.plot(white_noise,color="#FF914D")
plt.title(f"White Noise")
plt.subplot(2,1,2)
plt.plot(random_walk,color="#FF914D")
plt.title(f"random walk")
plt.tight_layout()
plt.show()

#ljung test for the whiter noise
ljung_test=acorr_ljungbox(white_noise,lags=[10],return_df=True)
print(ljung_test)


ljung_test=acorr_ljungbox(random_walk,lags=[10],return_df=True)
print(ljung_test)
```

#insights

# p>0.05 reject the null hypothesis else accept the alternative hypothesis
# in the first the white noise p value is greter than o.o5 that means there is no autocorrelstion
# in the random walk the p value is < 0.05 that means we accept the alternative hypothesis and reject the null hypotheisis that means here the striong autocorrelation exists
